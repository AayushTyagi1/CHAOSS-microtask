```python
from grimoirelabperceval.perceval.backends.core.git import Git
import datetime
import pprint
```

```python
Url = "https://github.com/AayushTyagi1/Music-Application"
Dir = "./app"
```

```python
git = Git(uri=Url,gitpath=Dir)
```

```python
for commit in git.fetch(category="commit"):
    pprint.pprint(commit["data"])
```

```python
range_commits = git.fetch()
range_commits_list = list(range_commits)
n = len(range_commits_list)
print("NUMBER OF COMMITS: ", n)
```

```python
last_commit = range_commits_list[n - 1]
print(last_commit)
```
